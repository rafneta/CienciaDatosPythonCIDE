#  Ciclos



### For

Un ciclo for es usado para iterare con respecto a una secuencia. Puede ser una lista, una tupla, un diccionario, un set o una cadena/string



Con el for podemos ejecutar un comando o una acción para cada elemento de una lista, set, diccionario, lista and so on.

for x in l3:
  print(x)

Podemos iterar también através de una cadena:

````python
saludar="Ciao Mamma"
for x in saludar:
    print(x)
````

Con el comando *break* podemos salir del *for* antes de iterare todos los objetos

````python
for x in saludar:
    print(x)
    if x== "m":
        break
````

Con el comando *continue* podemos parar la iteracción en curso del *for* y seguir con la siguiente

````python
Ciao Mamma

for x in saludar:
    if x== "m":
        continue
    print(x)
    
````

### While

El segundo tipo de ciclos que tenemos es el *while*. En este caso el codigo en le *while* se ejecutará hasta que una condición dada sea verdadera

````python
i = 1
while i < 6:
  print(i)
````

````python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
````

````python
grados = 115  
while grados > 112: # first while loop code
    print(grados)
    grados = grados - 1

print('Ahora es frio')
````

El ciclo procede hasta que la condición resultoó falsa.

## Loop Interactivo

Podemos hacer loop interactivos:

````python
lines = list() 
n = int(input('Cuantas lineas quieres ingrasar? ')) 
for i in range(n): 
    line = input('Linea: ') 
    lines.append(line) 
 
print('Tus lineas son:')  # check now 
for line in lines: 
    print(line) 
````

````python
lines = list()
testAnswer = input('Oprimir y si quieres introducir más lineas: ')
while testAnswer == 'y':
    line = input('Linea: ')
    lines.append(line)
    testAnswer = input('Oprimir y si quieres introducir más lineas: ')

print('Tus lineas fueron:')
for line in lines:
    print(line)
````

Los datos deben ser inicializados antes del loop para que funcione el test sobre la condición. El test de igual forma debe funcionar en el momento que el loop se repite. Por lo tanto tenemos que "preparar" los datos antes de acabar el ciclo.

````python
lines = list()
print('Ingresa una linea de codigo')
print('Enter an empty line to quit.')

line = input('Proxima linea: ') 
while line != '':           
    lines.append(line)
    line = input('Proxima linea: ')  

print('Tus linea: ')
for line in lines:
    print(line)
````

Siempre es mejor controlar dos vecves para averiguar que esta pasando a nuestra variable "de estado" al final del loop

Podemos tener distintas estructuras del loop, lo relevante es:
* pensar si hay alguna repetición en la operación que queremos hacer
* pensar en la condición que implica la continuación de la repetición del ciclo
* el resultado o la acción que queremos en cada ciclo
* pensar en el codigo suplementar necesario para luego estar listo par3a el siguiente ciclo

````python
from random import randint

random_number = randint(1, 10)
guesses_left = 3

while guesses_left>0:
    guess=input("Cual es el numero :")
    
    if guess==random_number:
        print ("Ganaste")
        break
    guesses_left-=1
else:
    print ("Perdiste")
````

````python
import re
p= input("Input your password")
x = True
while x:  
    if (len(p)<6 or len(p)>12):
        break
    elif not re.search("[a-z]",p):
        break
    elif not re.search("[0-9]",p):
        break
    elif not re.search("[A-Z]",p):
        break
    elif not re.search("[$#@]",p):
        break
    elif re.search("\s",p):
        break
    else:
        print("Valid Password")
        x=False
        break

if x:
    print("Not a Valid Password")
````

You are given a string and your task is to swap cases. In other words, convert all lowercase letters to uppercase letters and vice versa.

Example: Pythonist 2 → pYTHONIST 2

````python
def swap_case(s):
    st=str()
    for i in s:
        if(i.isupper()):
            st=st + i.lower()
        else:
            st= st +i.upper()
    return st
swap_case("Pythonist 2")
````
