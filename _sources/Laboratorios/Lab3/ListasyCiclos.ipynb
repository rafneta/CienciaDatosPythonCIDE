#  Listas y Ciclos

## Listas

### Tipos de estructuras de datos
La listas son utilizadas para almacenar multiples elementos en una sola variables.

En general en Python tenemos tres tipos de estructura de datos pre-construidas para almacenar los datos, cada una con su caracteristicas propias:

* Tuple: colección ordenable pero los elementos son inmutables. Se permiten duplicados
* Set: colección no ordenada y sin índice. No permite duplicados 
* Dictionary: colección no ordeda y mutable. No admite duplicados
* List: colección que es ordenable y mutable. Admite elementos duplicados

Para inicializar una lista es suficiente asignar a una variable una lista vacía por medio de dos brackets:

````python
a= []
type(a)
````

### Caracteristica elementos de una lista

Los elementos de una lista son:
* Ordenados: significa que una vez asignados a una lista los elementos tendrán un roden definido. Podemos acceder a ellos por medio de un indíce. En el momento que añadimos un elemento será colocado al final de la lista.
* Mutables: podemos añadir, quitar o cambiar elementos después que la lista haya ya sido creada.
* Admisibles de duplicados: dado que la lista es indicizada podemos tener elementos iguales

Ahora bien en una lista podemos poner distintos tipos de datos

````python
lista= [ "ciao", False, 4 , 4.2]
print(lista)
````


:::{admonition} Ciclo for
:class: tip

Un ciclo for es usado para iterare con respecto a una secuencia. Puede ser una lista, una tupla, un diccionario, un set o una cadena/string
:::


Es diferente a los for de otros lenguajes de prgramación. Se parece más a un metodo iterador que podemos encontrar en otros lenguajes de porgramación orientados a objectos. 

Con el for podemos ejecutar un comando o una acción para cada elemento de una lista, set, diccionario, lista and so on.

````python
for x in l3:
  print(x)
````

Podemos iterar también através de una cadena:

````python
saludar="Ciao Mamma"
for x in saludar:
    print(x)
````

Con el comando *break* podemos salir del *for* antes de iterare todos los objetos

````python
for x in saludar:
    print(x)
    if x== "m":
        break
````

Con el comando *continue* podemos parar la iteracción en curso del *for* y seguir con la siguiente

````python
for x in saludar:
    print(x)
    if x== "m":
        continue
    
````

Si queremos repetir un mismo codigo un determinado numero de veces podemos utilizar la función *range*. La función *range* nos retorna una secuencia de numeros iniciando del $0$ y cada vez aumentandolo de $1$

````python
for x in range(10):
  print(x)
````

````python
for x in range(2,10,2):
    print(x)
````

Podemos utilizar un *else* en el ciclo *for*

````python
for x in l3:
  print(x)
else:
  print("Salida")
````

````python
for x in l3:
  if x == 5: break
  print(x)
else:
  print("Salida")
````

Podemos anida varios *for*: 

````python
pal=["Abbiamo", "quasi", "finito"]
for x in range(3):
    print(x)
    for k in pal:
        print(k)
````

En caso que queramos hacer un *for* vacío es necesario inserir un *pass* para no tener error

````python
for x in range(5):
    pass
````

## While

El segundo tipo de ciclos que tenemos es el *while*. En este caso el codigo en le *while* se ejecutará hasta que una condición dada sea verdadera

````python
i = 1
while i < 6:
  print(i)
````

````python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
````

Podemo meter la condición *if* toda en una linea:

````python
num2=float(input("Inserir un numero: "))
'Par' if num2%2 == 0 else 'Impar'
````

````python
i = 1
while i < 6:
  print(i)
  print('ciao') if i == 3 else 3
  i += 1
````

Los *if* no tienen una sitnxi compacta tendriamos que adaptar la función *next*

````python
elem=[1,3,4,5,6,7,8,9,10]
res=next((x for x in elem if x%2==0), None)
res
````

You are given a string and your task is to swap cases. In other words, convert all lowercase letters to uppercase letters and vice versa.

Example: Pythonist 2 → pYTHONIST 2

````python
def swap_case(s):
    st=str()
    for i in s:
        if(i.isupper()):
            st=st + i.lower()
        else:
            st= st +i.upper()
    return st
swap_case("Pythonist 2")
````
