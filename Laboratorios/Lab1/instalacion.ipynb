
### Creación de entornos 

> Un entorno de anaconda es un directorio qeu cotinen una coleecióne specifica de paquetes y sus dependencias 

La creación de entornos es útil, por ejemplo, puede evita el conflicto entre actualizaciones de paquetes no compatibles, trabajar con versiones especificas de los paquetes de forma independiente, etc.

Las siguientes instrucciones crean un ambiente  y lo activan

```
conda create -n mientorno python=3.7
conda activate mientorno
```

Se explican las instrucciones anteriores: 

**_conda create_**. Es l ainstrucción deconda para crear el ambiente

**_-n_**. Es un argumento para indicar como se llamará el ambiente

**_mientorno_**. Es el nombre que se ha decido dar al ambiente

**_pythin =3.7_**. Este ambiente tendra instalado el nucleo de Python versión 3.7

**_conda activate mientorno_**. Instrucción para activar el entorno con nombre _mientorno_

Estre entorno solo tiene instaldo el nucleo, entonces si necesitamos realizar graficas, trabajar con arreglos, acceder a métodos numericos y realizar esto en un notebook, necesitamos instalar los paquetes correspondintes

```
conda install numpy scipy matplotlib jupyter
```

Se explica la instrucción anterior:

**_conda install_**. Es la instrucción para instarlar un paquete en el ambiente actualmente activo

**_numpy scipy matplotlib jupyter_**. Los cuatros pauqtees que se instalarán


Posiblemente se quiera instalar un paquete que se encuentre en un canal especifico.


```{epigraph}
Los canales de conda son ubicaciones (direcciones especificas en internet) donde los paquetes están almacenados
```


El canal _conda-forge_ contiene paquetes que podrían no estar en el canal pre establcecido de _conda_. Instalremos jupyterlab de este canal en el entorno (mientorno) activo actualmente  


```
conda install -c conda-forge jupyterlab
```

Se explica la instrucción anterior:

**_conda install_**. Es la instrucción para instarlar un paquete en el ambiente actualmente activo

**_-c_**. Es un argumento para indicar el nombre del canal 

**_conda-forge_**. Nombre del canal

**_jupyterlab_**. Paquete a instalar


Para desactivar el entorno activo, se ejecuta la siguiente instrucción

```
conda deactivate
```

conda list 
codna info -e
conda list

conda update matplotlib



Para elegir una IDE (integrated development environment) de Python existen diferentes criterios

- [Spyder](https://www.spyder-ide.org)
- [Visual Studio Code](https://code.visualstudio.com/)
- [Atom](https://atom.io/)
- [PyCharm Community Edition](https://www.jetbrains.com/pycharm/). Es necesario enlazar con Anaconda 
- [RStudio](https://rstudio.com/products/rstudio/download/). Es necesaria la configuración de la biblioteca _reticulate_ para enlazar con Anaconda
- [Jupyter](https://jupyter.org/install)


## Para ejecución en la nube

- [Google Colab](https://colab.research.google.com)
- [Binder](https://mybinder.org)
- [DataLore](https://datalore.io)
- [Microsoft Azure ML](https://docs.microsoft.com/en-us/azure/machine-learning/how-to-manage-workspace?tabs=python) 

### Google Colab

[Google Colab](https://colab.research.google.com) (Colaboratory) servicio en línea gratuito, conectado con tu cuenta de Google

En el siguente enlace hay más información al respecto [![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/notebooks/intro.ipynb)

### Binder

[Binder](https://mybinder.org) es un servicio gratuito que ejecuta Jupyter Notebooks en línea de.  

El siguiente es un ejemplo de un Binder, puede tardar algunos minutos la construcción el ambiente [![Binder](https://mybinder.org/badge_logo.svg)](https://mybinder.org/v2/gh/rafneta/IntroduccionProbabilidad/master)


## Referencias

- Documentación de Anaconda [enlace](https://docs.anaconda.com/) 

-
